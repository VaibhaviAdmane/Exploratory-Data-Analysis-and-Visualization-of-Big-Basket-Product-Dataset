```python
# BigBasket Products Dataset — EDA & Visualization

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("BigBasket Products.csv")
print(" Dataset loaded successfully!")
print("Shape:", df.shape)
```

```python
print("\nData Info:")
print(df.info())
```

```python
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
print("\nSummary Statistics (Numeric Columns):")
print(df.describe().T)
```

```python
# Data Cleaning
# Drop duplicates if any
df.drop_duplicates(inplace=True)
```

```python
# Fill missing string columns with 'Unknown'
for col in ['product', 'brand', 'description']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
```

```python
# Fill numeric NaN with median
for col in ['rating']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
print("\n Missing values handled.")
print("New shape:", df.shape)
```

```python
# 1️ Univariate Analysis
# Sale Price Distribution 
if 'sale_price' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df["sale_price"], bins=40, kde=True, color="skyblue")
    plt.title("Distribution of Sale Price")
    plt.xlabel("Sale Price")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Market Price Distribution
if 'market_price' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df["market_price"], bins=40, kde=True, color="salmon")
    plt.title("Distribution of Market Price")
    plt.xlabel("Market Price")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Rating Distribution
if 'rating' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df["rating"], bins=20, kde=True, color="lightgreen")
    plt.title("Distribution of Product Ratings")
    plt.xlabel("Rating")
    plt.ylabel("Count")
    plt.show()
```

```python
#  2️ Category-Level Insights
if 'category' in df.columns:
    plt.figure(figsize=(12, 6))
    df["category"].value_counts().head(10).plot(kind="bar", color="cornflowerblue")
    plt.title("Top 10 Product Categories")
    plt.xlabel("Category")
    plt.ylabel("Number of Products")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'sub_category' in df.columns:
    plt.figure(figsize=(12, 6))
    df["sub_category"].value_counts().head(10).plot(kind="bar", color="teal")
    plt.title("Top 10 Subcategories")
    plt.xlabel("Sub Category")
    plt.ylabel("Number of Products")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'brand' in df.columns:
    plt.figure(figsize=(12, 6))
    df["brand"].value_counts().head(10).plot(kind="bar", color="orchid")
    plt.title("Top 10 Brands by Number of Products")
    plt.xlabel("Brand")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 3️ Price vs Category / Brand
if 'category' in df.columns and 'sale_price' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x="category", y="sale_price", data=df, estimator=np.mean, ci=None)
    plt.title("Average Sale Price by Category")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'category' in df.columns and 'rating' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x="category", y="rating", data=df, estimator=np.mean, ci=None)
    plt.title("Average Rating by Category")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'sale_price' in df.columns and 'rating' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="sale_price", y="rating", data=df, alpha=0.6)
    plt.title("Relationship between Sale Price and Rating")
    plt.xlabel("Sale Price")
    plt.ylabel("Rating")
    plt.show()
```

```python
#Correlation Heatmap
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
if len(num_cols) > 1:
    plt.figure(figsize=(8, 6))
    sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# Brand vs Price 
if 'brand' in df.columns and 'sale_price' in df.columns:
    top_brands = df["brand"].value_counts().head(8).index
    df_top = df[df["brand"].isin(top_brands)]

    plt.figure(figsize=(12, 6))
    sns.boxplot(x="brand", y="sale_price", data=df_top, palette="Set3")
    plt.title("Sale Price Distribution among Top Brands")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# brand vs rating
if 'brand' in df.columns and 'rating' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x="brand", y="rating", data=df_top, ci=None, palette="viridis")
    plt.title("Average Rating among Top Brands")
    plt.xticks(rotation=45)
    plt.show()
```
